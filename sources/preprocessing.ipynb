## 3.1.&nbsp;Giá trị bị khuyết là biến kiểu số

```python
# Ở phần này, chúng tôi tiến hành thay thế những giá trị bị khuyết bằng mode cục bộ
# trong mỗi nhóm được tạo ra khi gom nhóm bởi thuộc tính 'Hãng'.
# Ngoài ra, tại mỗi nhóm nếu có giá trị nằm ngoài min và max của nhóm đó thì
# cũng thực hiện thay thế tương tự như trên.
```

```python
def get_group_min_max(df, groupby, column):
    """
    Lấy giá trị tối thiểu và tối đa trong mỗi nhóm
    được xác định bởi 'groupby' cho một 'column' cụ thể.

    Parameters:
    - df: DataFrame đầu vào.
    - groupby: Cột được sử dụng để nhóm.
    - column: Cột mà bạn muốn tìm giá trị tối thiểu và tối đa.

    Returns:
    Tuple của giá trị tối thiểu và tối đa trong mỗi nhóm.
    """

    non_nan_groups = df[df[column].notna()].groupby(groupby)[
        column].apply(list)
    x, y = non_nan_groups.apply(lambda x: stats.mode(
        x, keepdims=True)).apply([min, max])
    return x[0][0], y[0][0]


def handle_empty_group(df, groupby, column, strategy='replace_with_default', inplace=True):
    """
    Xử lý trường hợp nhóm có toàn bộ giá trị là NaN.

    Parameters:
    - df: DataFrame đầu vào.
    - groupby: Cột được sử dụng để nhóm.
    - column: Cột cần xử lý.
    - strategy: Chiến lược xử lý ('remove_group', 'replace_with_default', 'remove_column').
    - inplace: Nếu True, sửa đổi DataFrame trực tiếp; nếu False, trả về DataFrame mới.

    Returns:
    Nếu inplace=True, None. Nếu inplace=False, trả về DataFrame
    sau khi xử lý.
    """
    grouped = df.groupby(groupby)[column]

    # Kiểm tra nếu toàn bộ giá trị trong một nhóm đều là NaN
    empty_groups = grouped.apply(
        lambda x: x.notna().all()
    ).loc[lambda x: x == False].index

    if inplace:
        if strategy == 'remove_group':
            # Loại bỏ nhóm đó khỏi DataFrame
            df.drop(df[df[groupby].isin(empty_groups)].index, inplace=True)
        elif strategy == 'replace_with_default':
            # Thay thế giá trị NaN trong nhóm bằng giá trị mặc định hoặc giá trị thống kê tổng quát
            # Bạn có thể chọn giá trị khác tùy thuộc vào nhu cầu
            default_value = df[column].mode()[0]
            df[column].fillna(default_value, inplace=True)
        elif strategy == 'remove_column':
            # Loại bỏ cột đó khỏi DataFrame
            df.drop(column, axis=1, inplace=True)
    else:
        # Tạo một bản sao của DataFrame để tránh thay đổi trực tiếp
        df = df.copy()
        if strategy == 'remove_group':
            # Loại bỏ nhóm đó khỏi DataFrame
            df.drop(df[df[groupby].isin(empty_groups)].index, inplace=True)
        elif strategy == 'replace_with_default':
            # Thay thế giá trị NaN trong nhóm bằng giá trị mặc định hoặc giá trị thống kê tổng quát
            # Bạn có thể chọn giá trị khác tùy thuộc vào nhu cầu
            default_value = df[column].mode()[0]
            df[column].fillna(default_value, inplace=True)
        elif strategy == 'remove_column':
            # Loại bỏ cột đó khỏi DataFrame
            df.drop(column, axis=1, inplace=True)

    if not inplace:
        return df


def make_group_nan_and_fill_mode_numeric(df, groupby, column, replace_outliers=True, inplace=True, empty_group_strategy='replace_with_default'):
    """
    Gán giá trị NaN cho 'column' cho các giá trị nằm ngoài
    giá trị tối thiểu và tối đa thực tế của nhóm.
    Điền giá trị NaN bằng mode cục bộ trong mỗi nhóm.

    Parameters:
    - df: DataFrame đầu vào.
    - groupby: Cột được sử dụng để nhóm.
    - column: Cột cần xử lý.
    - replace_outliers: Nếu True, thay thế giá trị nằm ngoài min và max bằng NaN.
    - inplace: Nếu True, sửa đổi DataFrame trực tiếp; nếu False, trả về DataFrame mới.
    - empty_group_strategy: Chiến lược xử lý cho nhóm toàn bộ giá trị là NaN ('remove_group', 'replace_with_default', 'remove_column').

    Returns:
    Nếu inplace=True, None. Nếu inplace=False, trả về DataFrame
    với giá trị NaN đã được gán và điền.
    """

    # Lấy giá trị tối thiểu và tối đa trong mỗi nhóm
    # Được xác định bởi 'groupby' cho một 'column' cụ thể.
    mini, maxi = get_group_min_max(df, groupby, column)

    # Gán giá trị sai thành NaN nếu replace_outliers=True
    if replace_outliers:
        col = df[column].apply(
            lambda x: np.NaN if not (mini <= x <= maxi) else x)
    else:
        col = df[column]

    # Điền bằng mode cục bộ
    mode_by_group = df.groupby(groupby)[column].transform(
        lambda x: x.mode()[0] if not x.mode().empty else np.NaN)

    # Kết hợp với xử lý nhóm toàn bộ giá trị là NaN
    df[column] = col.fillna(mode_by_group)

    # Xử lý nhóm toàn bộ giá trị là NaN
    handle_empty_group(df, groupby, column,
                       strategy=empty_group_strategy, inplace=inplace)

    if not inplace:
        return df


def numeric_wrong_values_reassign_group_min_max(df, groupby, column, replace_outliers=True, inplace=True, empty_group_strategy='replace_with_default'):
    """
    Gán giá trị NaN cho 'column' cho các giá trị nằm ngoài giá trị tối thiểu và tối đa thực tế của nhóm.
    Điền giá trị NaN bằng mode cục bộ trong mỗi nhóm.
    Tùy chọn hiển thị các thay đổi trước và sau khi gán lại.

    Parameters:
    - df: DataFrame đầu vào.
    - groupby: Cột được sử dụng để nhóm.
    - column: Cột cần xử lý.
    - replace_outliers: Nếu True, thay thế giá trị nằm ngoài min và max bằng NaN.
    - inplace: Nếu True, sửa đổi DataFrame trực tiếp; nếu False, trả về DataFrame mới.
    - empty_group_strategy: Chiến lược xử lý cho nhóm toàn bộ giá trị là NaN ('remove_group', 'replace_with_default', 'remove_column').

    Returns:
    Nếu inplace=True, None. Nếu inplace=False, trả về DataFrame
    với giá trị NaN đã được gán và điền.
    """

    if inplace:
        print(f"--- Thay Thế Giá Trị Không Hợp Lệ Cho Cột {column} ---")

        # Trước khi Gán giá trị NaN
        nan_values_before = df[column].value_counts(
            dropna=False).get(np.NaN, 0)
        if nan_values_before:
            print(f'\nTrước khi Gán: {column}',
                  f'Có {nan_values_before} Giá Trị NaN', end='\n')

        if replace_outliers:
            actual_min, actual_max = get_group_min_max(df, groupby, column)
            print(f"\nMin Và Max Ban Đầu Của Thuộc Tính:", df[column].apply(
                [min, max]), sep='\n', end='\n')

            print(f"\nMin Và Max Thực Tế Khi Thực Hiện Gom Nhóm Bởi {groupby}:",
                  f'Min\t{actual_min},\nMax\t{actual_max}', sep='\n', end='\n')

        example_before = df.groupby(groupby)[column].apply(list).head()
        print(f'\nVí Dụ Trước Khi Gán {column}:',
              *example_before.values, sep='\n', end='\n')

        # Gán giá trị
        make_group_nan_and_fill_mode_numeric(
            df, groupby, column, replace_outliers, inplace)

        # Sau khi Gán giá trị NaN
        nan_values_after = df[column].value_counts(dropna=False).get(np.NaN, 0)
        if nan_values_after:
            print(f'\nSau Khi Gán: {column}',
                  f'Có {nan_values_after} Giá Trị NaN', end='\n')

        example_after = df.groupby(groupby)[column].apply(list).head()
        print(f'\nVí Dụ Sau Khi Gán {column}:',
              *example_after.values, sep='\n', end='\n')

        # Xử lý nhóm toàn bộ giá trị là NaN
        handle_empty_group(df, groupby, column,
                           strategy=empty_group_strategy, inplace=True)

    else:
        # Trả về DataFrame với giá trị NaN đã được gán và điền
        result = make_group_nan_and_fill_mode_numeric(
            df, groupby, column, replace_outliers, inplace)

        # Xử lý nhóm toàn bộ giá trị là NaN
        handle_empty_group(result, groupby, column,
                           strategy=empty_group_strategy, inplace=False)

        return result
```

## 3.2. Giá trị bị khuyết là biến phân loại

```python
def fill_nan_with_most_common_per_group(df, column_name, groupby_column='Customer_ID'):
    """
    Điền giá trị NaN trong một cột của DataFrame bằng giá trị phổ biến nhất không phải NaN
    trong mỗi nhóm được xác định bởi một cột cụ thể.

    Tham số:
    - df: DataFrame đầu vào.
    - column_name: Cột cần xử lý.
    - groupby_column: Cột được sử dụng để nhóm. Mặc định là 'Customer_ID'.

    Trả về:
    None. Hàm thực hiện sửa đổi trực tiếp trên DataFrame đầu vào.
    """

    # Bước 1: Gom nhóm DataFrame theo cột được chỉ định
    # Và tìm giá trị không phải NaN phổ biến nhất cho mỗi nhóm
    most_common_values = df.groupby(groupby_column)[column_name].agg(
        lambda x: x.mode()[0] if pd.notna(x).any() else np.NaN
    )

    # Bước 2: Thay thế giá trị NaN bằng giá trị phổ biến nhất cho mỗi nhóm
    df[column_name] = df.apply(
        lambda row: most_common_values[row[groupby_column]]
        if pd.isna(row[column_name]) else row[column_name],
        axis=1
    )

    # Bước 3: Xử lý nhóm toàn bộ giá trị là NaN
    handle_empty_group_categories(df, groupby_column, column_name)


def handle_empty_group_categories(df, groupby_column, column_name):
    """
    Xử lý nhóm toàn bộ giá trị là NaN trong DataFrame.

    Tham số:
    - df: DataFrame đầu vào.
    - groupby_column: Cột được sử dụng để nhóm.
    - column_name: Cột cần xử lý.

    Trả về:
    None. Hàm thực hiện sửa đổi trực tiếp trên DataFrame đầu vào.
    """

    # Kiểm tra nếu toàn bộ giá trị trong một nhóm đều là NaN
    empty_groups = df.groupby(groupby_column)[column_name].apply(
        lambda x: x.notna().all()
    ).loc[lambda x: x == False].index

    # Xử lý nhóm toàn bộ giá trị là NaN
    df.loc[df[groupby_column].isin(empty_groups), column_name] = df[column_name].mode()[
        0] if not df[column_name].mode().empty else np.NaN
```

```python
# Thay thế các giá trị NaN trong biến phân loại bằng mode cục bộ của từng nhóm khi gom nhóm
# bởi 'Hãng'. Nếu nhóm đó rỗng thì thay thế bằng mode toàn cục của cột đang xét.
fill_nan_with_most_common_per_group(df, 'Phân loại người bán', 'Hãng')
fill_nan_with_most_common_per_group(df, 'Kiểu dáng', 'Hãng')
```

## 5.3.&nbsp;Bộ dữ liệu dùng One Hot Encoder

```python
# Tạo một bản sao của DataFrame df và lưu vào df_onehot
df_onehot = df.copy()

# Áp dụng One-Hot Encoding cho các cột dạng categorical trong DataFrame
df_onehot = pd.get_dummies(df_onehot, prefix=categorical_columns)

# Lấy cột 'Giá' và lưu vào biến 'price'
price = df_onehot.pop('Giá')

# Thêm cột 'Giá' vào cuối DataFrame
df_onehot['Giá'] = price
```
